#### Coming back to bs4


```python
source=urllib.request.urlopen(url)
```

```python
soup= bs.BeautifulSoup(source,'lxml')
```

```python
text=''
for paragraph in soup.find_all('p'):
    text+= paragraph.text
         
```

```python
 # text need preprocessing 
text=re.sub(r"\[[0-9]*\]"," ",text)
```

```python
text=re.sub(r'\s+',' ',text)
```

```python
clean_text=text.lower()
clean_text=re.sub(r'\W',' ',clean_text)
clean_text=re.sub(r'\d',' ',clean_text)
clean_text=re.sub(r'\s+',' ',clean_text)
                  
```

```python
import pandas as pd
import nltk
```

```python
sentance=nltk.sent_tokenize(text)
```

```python
sentance
```

```python
df=pd.DataFrame(sentance)
```

```python
df.head(15)
```

```python
stopwords=nltk.corpus.stopwords.words('english')
```

```python
stopwd=pd.DataFrame(stopwords)
```

```python
stopwd.head(15)
```

```python
wordCount={}
for word in nltk.word_tokenize(clean_text):
    if word not in stopwords:
        if word not in wordCount.keys():
            
            wordCount[word]=1
        else:
            wordCount[word]+=1
```

```python
wordCount
```

```python
for key in wordCount.keys():
    wordCount[key]=wordCount[key]/max(wordCount.values())
```

```python
wordCount
```

```python
'''import matplotlib.pyplot as plt
plt.scatter(wordCount,wordCount)
plt.show()'''
```

```python
sentanceScore={}
for sent in sentance:
    for word in nltk.word_tokenize(sent.lower()):
        if word in wordCount.keys():
            if len(sent.split(' '))<30:
                if sent not in sentanceScore.keys():
                    sentanceScore[sent]=wordCount[word]
                else:
                    sentanceScore[sent]+=wordCount[word]
```

```python
import heapq
```

```python
best_Summary=heapq.nlargest(5,sentanceScore,key=sentanceScore.get)
```

```python
best_Summary
```

```python
best_Summary1=pd.DataFrame(best_Summary)
best_Summary1
```
